```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = 0.001
t = np.arange(0,1,dt)
f0 = 50
f1 = 350
t1 = 2
x = np.cos(2*np.pi*t*(f0 + (f1-f0)*np.power(t,2)/(3*t1**2)))
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.specgram(x, NFFT=2**7, Fs=1/dt, noverlap=120)
plt.colorbar()
```

```python
fhat = np.fft.fft(x)
psd = np.real(fhat*np.conj(fhat))
freq = (1/(dt*len(t))) * np.arange(len(t)) 
sns.lineplot(x=freq[:len(t)//2], y=psd[:len(t)//2])
```
